```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = r"./IMG_20210722_152010.jpg"
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, blackAndWhiteImage) = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.imshow(blackAndWhiteImage)
```

```python
def GreyScale(path):
  img = cv2.imread(path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  img = cv2.resize(img, (500, 500))
  H,W = img.shape[:2]
  gray = np.zeros((H,W), np.uint8)
  for i in range(H):
      for j in range(W):
          gray[i,j] = np.clip(0.07 * img[i,j,0]  + 0.72 * img[i,j,1] + 0.21 * img[i,j,2], 0, 255)
  return gray
```

```python
path = r"./IMG_20210722_152010.jpg"
gray = GreyScale(path)
plt.imshow(gray)
```

```python
def BlackAndWhite(img):
  w = 120
  for y, line in enumerate(img):
    for x, color in enumerate(line):
      if color >= w:
        img[y][x] = 255
      else:
        img[y][x] = 0
  
  return img
```

```python
baw = BlackAndWhite(img)
```

```python
plt.imshow(baw)
```
